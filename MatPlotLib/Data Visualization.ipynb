# Aula 50 Implementing Figures and Axis

```python
a = np.linspace(0, 10, 11)
```

```python
b = a**4
```

```python
x = np.arange(0, 10)
```

```python
y = 2*x
```

```python
fig = plt.figure()

# LARGE AXES
axes1 = fig.add_axes([0,0,1,1])
axes1.plot(a, b)

# SMALL AXES
axes2 = fig.add_axes([0.2,0.2, 0.5, 0.5])
axes2.plot(a,b)
```

```python
fig = plt.figure(figsize=(10,3), dpi=300)
axes1 = fig.add_axes([0,0,1,1])
axes1.plot(a, b)
```

```python
fig, axes = plt.subplots(nrows = 2, ncols=2, dpi=300)
axes[0][0].plot(x, y)
axes[1][1].plot(a, b)


fig.suptitle("TITLE", fontsize=16)
fig.tight_layout()
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x, x, label='X vs X', color='orange')
ax.plot(x, x**4, label='X vs X^2', color='yellow')
ax.legend(loc='lower left')
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x, x, label='X vs X', lw=5)
ax.plot(x, x**4, label='X vs X^2', color='yellow', ls=':', lw=3)
ax.legend(loc='lower left')
```
