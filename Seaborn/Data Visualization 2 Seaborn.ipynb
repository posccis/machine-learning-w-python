```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
```

```python
df = pd.read_csv('.\\UNZIP_FOR_NOTEBOOKS_FINAL\\05-Seaborn\\dm_office_sales.csv')
```

```python
df.head()
```

# ScatterPlot 

```python
plt.figure(figsize=(12,4))
sns.scatterplot(x='salary', y='sales', data=df, s=200, hue='level of education',
                style='level of education')
plt.show()
```

# Distribuition Plot

```python
sns.rugplot(x='salary', data=df, height=0.5)
```

```python
sns.set(style='ticks')
sns.displot(data=df, x='salary', kde=True)
```

```python
sns.kdeplot(data=df, x='salary', bw_adjust=0.1)
```

# Categorical Plots

```python
plt.figure(figsize=(10,4), dpi=200)
sns.countplot(data=df, x='level of education', hue='division', palette='Set1')
```

```python
plt.figure(figsize=(10,4), dpi=200)
sns.barplot(data=df, x='level of education', y='salary', estimator=np.mean, ci='sd', hue='division', palette='Set1')
```

```python
df = pd.read_csv('.\\UNZIP_FOR_NOTEBOOKS_FINAL\\05-Seaborn\\StudentsPerformance.csv')
```

```python
df.head()
```

```python
plt.figure(figsize=(10,4), dpi=200)
sns.boxplot(data=df, y='reading score', x='parental level of education',
           hue='test preparation course')
plt.legend(bbox_to_anchor=(1.05, 1))
```

```python
plt.figure(figsize=(10, 8), dpi=200)

sns.violinplot(data=df, x='reading score', y='parental level of education',
               hue='test preparation course', palette='Set1', split=True)
```

```python
plt.figure(figsize=(8,4), dpi=200)
sns.swarmplot(data=df, x='math score', y='gender',
              hue='test preparation course', size=3)
plt.legend(bbox_to_anchor=(1.05, 0.5))
```

```python
sns.boxenplot(x='math score', y='test preparation course', data=df, hue='gender')
```

# Comparison Plots

```python
sns.jointplot(data=df, x='math score', y='reading score', hue='gender')
```

```python
sns.pairplot(data=df, hue='gender')
```

```python
sns.catplot(data=df, x='gender', y='math score', kind='box',
           col='lunch')
```

```python
g = sns.PairGrid(df, hue='gender')

g = g.map_upper(sns.scatterplot)

g = g.map_diag(sns.histplot)

g = g.map_lower(sns.kdeplot)
```
